# Comparison of Anomaly Detection method

Current implementation uses a z-score detector. The model starts off with an initial “calibration” phase where it collects data to build a distribution. This allows us estimate the parameters of the normal distribution. Once in detection mode, we can continuously compare new data to the distribution to how far it deviates from the mean in terms of standard deviation for classification. New data is standardized into a z-score and compared to a threshold we set (2 or 3(stricter))

```python
from scipy import signal
def plot_spectrogram(sample, fs, title):
    f, t, Sxx = signal.spectrogram(sample, fs)
    plt.figure(figsize=(12, 8))
    plt.pcolormesh(t, f, 10 * np.log10(Sxx), shading='gouraud')
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]')
    plt.title(title)
    plt.colorbar(label='Intensity [dB]')
    plt.show()

fs = 1000  # Replace with your actual sampling frequency
plot_spectrogram(X_train[0], fs, 'Spectrogram of First Sample')
```

```python
from statsmodels.tsa.stattools import adfuller
from tqdm import tqdm

def adf_test(series):
    result = adfuller(series.dropna())
    return pd.Series({
        'ADF Statistic': result[0],
        'p-value': result[1],
        'Is Stationary': result[1] < 0.05
    })

def test_stationarity_by_group(data):
    # Separate features and labels
    features = data.iloc[:, :-1]
    labels = data.iloc[:, -1]
    
    # Group by label
    grouped = pd.concat([features, labels], axis=1).groupby(labels.name)
    
    results = {}
    for name, group in grouped:
        print(f"\nTesting stationarity for group {name}")
        group_features = group.iloc[:, :-1]
        
        group_results = []
        for column in tqdm(group_features.columns, desc=f"Group {name}"):
            result = adf_test(group_features[column])
            result['Frequency Bin'] = column
            group_results.append(result)
        
        group_results_df = pd.DataFrame(group_results).set_index('Frequency Bin')
        results[name] = group_results_df
        
        # Print summary for this group
        stationary_count = sum(group_results_df['Is Stationary'])
        total_count = len(group_results_df)
        print(f"Group {name} Summary:")
        print(f"Stationary Frequency Bins: {stationary_count}/{total_count} ({stationary_count/total_count*100:.2f}%)")
    
    return results

results = test_stationarity_by_group(data)

# Print detailed results for the first few frequency bins of each group
for name, result in results.items():
    print(f"\nDetailed results for first few frequency bins of group {name}:")
    print(result.head())
```
